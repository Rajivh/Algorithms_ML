```python
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


def linear_regression(training, features, output, theta, learning_rate, precision):
    beta_0 = theta[0]
    beta_1 = theta[1]
    cost_array = []
    beta_0_array = []
    beta_1_array = []
    cost_prev = 0
    cost_curr = 1
    
    while(abs(cost_curr - cost_prev) > precision):
    #for i in range(iterations):
        print("beta_1",beta_1)
        cost_prev = cost_curr
        hyp = beta_0 + (beta_1*features)
        #hyp = (beta_1*features)
        error = sum((hyp - output)**2)
        cost_curr = error/(2*training)
        cost_array.append(cost_curr)
        beta_0_array.append(beta_0)
        beta_1_array.append(beta_1)
        
        beta_0_gradient = sum(hyp - output)
        beta_1_gradient = sum((hyp - output)*features)
        beta_0 = beta_0 - (learning_rate/training)*beta_0_gradient
        beta_1 = beta_1 - (learning_rate/training)*beta_1_gradient
        
        
    print(cost_array)
    #print(beta_1_array)
    plot_function(beta_0_array,beta_1_array,cost_array)
  
        



```

```python
def plot_function(beta_0,beta_1,cost):
    #print(beta_1)
    plt.plot(beta_1,cost)
    plt.suptitle('Cost vs parameter', fontsize=14, fontweight='bold')
    plt.xlabel('Theta1')
    plt.ylabel('Cost')

    plt.show()
```

```python
def plot_scatter(features, output):
    plt.scatter(features, output, alpha=0.4)
    plt.xlabel('Age')
    plt.ylabel('Income')
    plt.show()
```
